```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset
df = pd.read_csv("Kelas F_Student AI Usage.csv")

# Menghitung rata-rata study hours
avg_study = df["study_hours_per_day"].mean()

# Memfilter data:
# study_hours_per_day di atas rata-rata dan tidak menggunakan AI
filtered = df[
    (df["study_hours_per_day"] > avg_study) &
    (df["uses_ai"] == "No")
]

# Membuat grafik 2
plt.figure(figsize=(8, 5))

sns.barplot(
    data=filtered,
    x=filtered.index,
    y="study_hours_per_day"
)

plt.title("Above Average Study Hours & No AI")
plt.xlabel("Student")
plt.ylabel("study_hours_per_day")

plt.tight_layout()
plt.show()
```
