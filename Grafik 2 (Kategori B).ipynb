```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data (pastikan file CSV berada di folder yang sama dengan script)
df = pd.read_csv('Kelas F_Student AI Usage.csv')

# 2. Hitung rata-rata jam belajar keseluruhan
avg_study = df['study_hours_per_day'].mean()

# 3. Filter data sesuai kondisi
filtered_df = df[(df['study_hours_per_day'] > avg_study) & (df['uses_ai'] == 'No')].reset_index()

# 4. Membuat Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(filtered_df.index + 1, filtered_df['study_hours_per_day'], color='royalblue', edgecolor='black', alpha=0.8)

# 5. Menambahkan garis rata-rata keseluruhan sebagai acuan
plt.axhline(y=avg_study, color='red', linestyle='--', linewidth=1.5, label=f'Rata-rata Keseluruhan ({avg_study:.2f} Jam)')

# 6. Menambahkan label angka di atas setiap bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f"{yval} Jam", ha='center', va='bottom', fontsize=9, fontweight='bold')

# 7. Mengatur Kelengkapan Grafik (Judul, Axis, Legend)
plt.title('Siswa dengan Jam Belajar di Atas Rata-rata (Tanpa AI)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Siswa Ke-', fontsize=12)
plt.ylabel('Jam Belajar per Hari', fontsize=12)
plt.xticks(filtered_df.index + 1)
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.legend(loc='upper right')

# 8. Tampilkan Grafik
plt.tight_layout()
plt.show()
```
