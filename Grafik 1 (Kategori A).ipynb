```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
# Pastikan file CSV berada di folder yang sama dengan file script ini
file_path = 'Kelas F_Student AI Usage.csv'
df = pd.read_csv(file_path)

# 2. Pengolahan Data
# Mengelompokkan berdasarkan education_level dan mencari nilai max screen time
data_grafik = df.groupby('education_level')['daily_screen_time_hours'].max().sort_values(ascending=False).reset_index()

# 3. Pengaturan Visualisasi
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid") # Membuat background grafik lebih bersih

# Membuat Bar Chart
# Palette 'viridis' atau 'rocket' memberikan gradasi warna yang bagus
ax = sns.barplot(
    x='education_level', 
    y='daily_screen_time_hours', 
    data=data_grafik, 
    palette='viridis'
)

# 4. Menambahkan Label Angka di Atas Bar (Agar data mudah dibaca)
for p in ax.patches:
    ax.annotate(f'{p.get_height()} jam', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 9), 
                textcoords='offset points',
                fontsize=11,
                fontweight='bold')

# 5. Menambahkan Judul dan Keterangan Sumbu
plt.title('Maksimum Durasi Layar Harian Berdasarkan Tingkat Pendidikan', fontsize=14, pad=20)
plt.xlabel('Tingkat Pendidikan', fontsize=12)
plt.ylabel('Durasi Layar (Jam)', fontsize=12)

# 6. Menampilkan Grafik (PENTING: Baris ini yang memunculkan jendela grafik)
plt.tight_layout()
plt.show()
```
