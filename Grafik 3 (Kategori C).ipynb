```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset
df = pd.read_csv("Kelas F_Student AI Usage.csv")

# Membuat kolom baru
df["Grade_Improvement"] = df["grades_after_ai"] - df["grades_before_ai"]

# Scatter Plot
fig, ax = plt.subplots(figsize=(8, 6))

sns.scatterplot(
    data=df,
    x="study_hours_per_day",
    y="Grade_Improvement",
    ax=ax
)

ax.set_title("Hubungan Waktu Belajar Harian dengan Peningkatan Nilai")
ax.set_xlabel("Waktu Belajar per Hari (Jam)")
ax.set_ylabel("Peningkatan Nilai (Grade Improvement)")

plt.show()
```
